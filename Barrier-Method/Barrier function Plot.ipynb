```julia
using JuMP, Ipopt
function plot_minimizer(t)

    m = Model(solver = IpoptSolver())
    @variable(m,1 <= x <= 2)
    @NLobjective(m, Min, x^2 - t*log(-(x-2)) - t*log(-(-x+1)))
    solve(m);
    xopt = getvalue(x)
    scatter([xopt], f([xopt],t))
end
function plot_minimizer_barrier(t)

    m = Model(solver = IpoptSolver())
    @variable(m,1 <= x <= 2)
    @NLobjective(m, Min,  - t*log(-(x-2)) - t*log(-(-x+1)))
    solve(m);
    xopt = getvalue(x)
    scatter([xopt], f([xopt],t))
end
```

```julia
using PyPlot

 = 0.0000001
x = linspace(1+,2-,1000)
y = x
f_1(x) = x-2
f_2(x) = -x +1

f(X,t) = [x^2 - t*log(-f_1(x)) - t*log(-f_2(x)) for x in X]

for t in [5,2,1,0.5,0.1]
    plot(x,f(x,t), label = "t = $t")
    plot_minimizer(t)
    
end
x = linspace(-2,3,1000)
y = x.^2
plot(x,y,label = L"$f(x) = x^2$")

title(L"Barier function for $f(x) = x^2$ with $1 \leq x \leq 2$")
xlim(-2,3);
ylim(0,20);
legend()


```
