```julia
using JuMP
using GLPKMathProgInterface
using Clp
using Ipopt
```

# Problem 2

```julia
m = Model(with_optimizer(Clp.Optimizer))


@variable(m, x_1 >= 0)
@variable(m, x_2 >= 0)

@objective(m, Min, -x_1 - 3x_2)
@constraint(m, 2x_1 + 3x_2 <= 6)
@constraint(m, -x_1 + 4x_2 <= 4)

display(m)
optimize!(m)
println(JuMP.objective_value(m))
println(JuMP.value(x_1))
println(JuMP.value(x_2))
```

# Problem 3

```julia
m = Model(with_optimizer(Ipopt.Optimizer))

@variable(m, x_1)
@variable(m, x_2)

@NLobjective(m, Min, x_1^2 + 2x_2^2)
@constraint(m, x_1 + x_2 -2 == 0)

display(m)
optimize!(m)
println(JuMP.objective_value(m))
println(JuMP.value(x_1))
println(JuMP.value(x_2))
```

# Problem 4

```julia
m = Model(with_optimizer(Ipopt.Optimizer))

@variable(m, x_1)
@variable(m, x_2)

@objective(m, Min, x_1 + 2x_2)
@NLconstraint(m, (x_1)^2 + (x_2)^2 <= 1)

display(m)
optimize!(m)
println(JuMP.objective_value(m))
println(JuMP.value(x_1))
println(JuMP.value(x_2))
```

# Problem 5

```julia
m = Model(with_optimizer(Ipopt.Optimizer))

@variable(m, -3 <= x_1 <= 3)
@variable(m, -3 <= x_2 <= 3)

@NLobjective(m, Min, sin(x_1)*cos(2x_2) + sin(2x_1*x_2))

display(m)
optimize!(m)
println(JuMP.objective_value(m))
println(JuMP.value(x_1))
println(JuMP.value(x_2))    
```
